```python
# import neccessary Library

import pandas as pd
```

```python
# Load the Dataset

Dataset = pd.read_csv("CKD.csv")

Dataset
```

```python
# We need to encode the Categorical data using One-Hot-Encoder

Dataset = pd.get_dummies(Dataset, dtype = int, drop_first = True)

Dataset
```

```python
Dataset.columns
```

```python
# We are checking whether our target column is Balanced or Imbalanced

Dataset["classification_yes"].value_counts()
```

```python
# We need to know the column names and its spellings correctly

Dataset.columns
```

```python
# We are splitting the Input and Output Variable

Independent = Dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]

Independent
```

```python
Dependent = Dataset[["classification_yes"]]

Dependent
```

```python
# Now we are splitting Train and Test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Independent, Dependent, test_size = 0.2, random_state = 0)
```

```python
# Now we are going to satndardise our input variables

from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
x_train = SC.fit_transform(x_train)
x_test = SC.transform(x_test)
```

```python
# Now we need to import GRidSearchCV and select the algorithm

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
             'gamma':['auto','scale'],
             'C':[10,100,1000,2000,3000]} 
grid = GridSearchCV(SVC(probability = True), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
```

```python
# Applying the formula

grid.fit(x_train, y_train) 
```

```python
# To check our Result
Result = grid.cv_results_

Result
```

```python
# To view the above result in Tabular form

Table = pd.DataFrame.from_dict(Result)

Table
```

```python
# We tested using test set
grid_pred = grid.predict(x_test)
```

```python
# Now we need to Evaluate

from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test, grid_pred)
matrix
```

```python
# We need to prepare the report

from sklearn.metrics import classification_report
report = classification_report(y_test,grid_pred)
print(report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
